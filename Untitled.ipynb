```python
import pandas as pd
import numpy as np
df = pd.read_csv('AnimeWorld.csv')

print("Первые 5 строк датафрейма:")
display(df.head())
print("\n Случайная выборка из 5 строк:")
display(df.sample(5, random_state = 25))
print("Последние 5 строк датафрейма:")
display(df.tail())
print("\n Атрибуты датафрейма:")
print(df.columns.tolist())
print("\n Информация о датафрейме:")
print(df.info())
print("\n Типы данных")
print(df.dtypes)
print("\nКоличество ненулевых (non-null) значений:")
print(df.count())
print("\nИспользование памяти каждым столбцом (в байтах):")
print(df.memory_usage(deep=True))

print('''При беглом осмотре выявлены следующие проблемы:
 - Столбец date_added имеет тип object, но это дата, и его следует преобразовать в формат даты
 - Большое количество пропусков в столбцах: director, cast и country. Это может привести к неправильным выводам и искажению результатов анализа''')
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')

print("\nНаименования столбцов после приведения к нижнему регистру:")
print(df.columns)
print("\nПервые 5 строк после приведения к нижнему регистру:")
display(df.head())
```

```python
# Проверка на пропуски
missing_values = df.isnull().sum()
print("\nПропущенные значения в каждом столбце:")
print(missing_values[missing_values > 0])
```

```python
df = df.fillna('null')
print("\nПервые 5 строк после замены пропусков на 'null':")
display(df.head())

df['year'] = pd.to_datetime(df['year'], errors='coerce', format='%b %d, %Y')
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df.info()
```

```python
# Фильтрация с использованием метода query()
print('\n Фильтр по рейтингу больше 7')
filtered_query_1 = df.query("rating > 8").head()
display(filtered_query_1)

print('\n Фильтр строк, содержащих названия фильмов, которые начинаются с A или заканчиваются на e.')
filtered_query_2 = df.query("anime.str.startswith('A') or anime.str.endswith('e')").head()
display(filtered_query_2)

print('\n Фильтрация по жанру и рейтингу, большей 7\n')
filtered_query_3 = df.query("genre == 'Romance' and rating > 7").head()
display(filtered_query_3)

print('\n фильмы вышедшие после 2020 года от студии OLM)\n')
filtered_query_4 = df.query("year > 2020 and studio == 'OLM'").head()
display(filtered_query_4)

print('\n фильмы 2000 года в жанре Comedy )\n')
filtered_query_5 = df.query("year == 2020 and genre == 'Comedy'").tail()
display(filtered_query_5)
```

```python
#Фильтрация с использованием метода where()
print("\n Фильмы в жанре Drama после 2021 года ")
filtered_where_1 = df.where((df['genre'] == 'Drama') & (df['year'] < 2021)).dropna(how='all')
display(filtered_where_1.head())

print("\n Смешные фильмы от студии OLM:")
filtered_where_2 = df.where((df['genre'] == 'Comedy') & (df['studio'] == 'OLM')).dropna(how='all')
display(filtered_where_2.head())

print("\n Фильмы от студии Bones с рейтингом от 6 до 8:")
filtered_where_3 = df.where((df['studio'] == 'Bones') & (df['rating'].between(5, 8))).dropna(how='all')
display(filtered_where_3.head())

print("\n Романтические фильмы от студии ILCA с рейтингом больше 8:")
filtered_where_4 = df.where((df['genre'] == 'Romance') & (df['studio'] == 'ILCA') & (df['rating'] >= 8)).dropna(how='all')
display(filtered_where_4.head())

print("\n Романтические фильмы от студии ILCA с рейтингом больше 8:")
filtered_where_4 = df.where((df['genre'] == 'Romance') | (df['studio'] == 'ILCA') & (df['rating'] >= 8)).dropna(how='all')
display(filtered_where_4.head())
```

```python
rating = df['rating']

# Рассчитываем основные статистические показатели
mean_values = rating.mean()
median_values = rating.median()
mode_values = rating.mode().iloc[0] if not rating.mode().empty else None # Берем первую моду, если их несколько
std_values = rating.std()
describe_values = rating.describe()

# Квартильное распределение
quantiles = rating.quantile([0.25, 0.5, 0.75])

# Вывод результатов
print("Средние значения:", mean_values)
print("\nМедианы:", median_values)
print("\nМоды:", mode_values)
print("\nСтандартные отклонения:", std_values)
print("\nКраткое описание (describe):", describe_values)
print("\nКвартильное распределение:", quantiles)
```

```python
# Визуализация матрицы корреляций
df['description_len'] = df['anime'].str.len()

# Построение матрицы корреляций
numeric_cols = df[['rating', 'description_len']]
corr_matrix = numeric_cols.corr()

# Вывод матрицы корреляций
print("Матрица корреляций числовых признаков:")
display(corr_matrix)

# Выявление признаков с высокой корреляцией
high_correlation = corr_matrix[(corr_matrix > 0.7) | (corr_matrix < -0.7)]
display("Признаки с высокой корреляцией:", high_correlation)

print('''Если будет обнаружена высокая положительная корреляция между description_len и Рейтинг, это может указывать на то, 
что более длинные описания привлекают больше внимания зрителей и, следовательно, влияют на их оценку. 
Это может побудить студии и продюсеров создавать более детализированные и увлекательные описания для своих аниме, 
чтобы повысить интерес аудитории.
Зная, что длина описания может влиять на рейтинг, 
компании могут использовать это знание для прогнозирования успеха новых аниме, 
основываясь на характеристиках их описаний.''')
```

```python
# Пример 1: Таблица сопряженности для жанра и студии
genre_studio_crosstab = pd.crosstab(df['genre'], df['studio'], margins=True, margins_name="Итого")
print("Таблица сопряженности для Жанр и Студия:")
display(genre_studio_crosstab)

# Рассчитываем долю каждой категории
genre_studio_percentage = genre_studio_crosstab.div(genre_studio_crosstab.loc["Итого"], axis=1) * 100
print("\nДоля каждой категории (в %):")
display(genre_studio_percentage)
```

```python
# Пример 2: Таблица сопряженности для года выпуска и рейтинга
# Преобразуем рейтинг в категориальные значения (например, "низкий", "средний", "высокий")
bins = [0, 3, 6, 10]  # Определяем границы для категорий
labels = ['Низкий', 'Средний', 'Высокий']
df['rating_category'] = pd.cut(df['rating'], bins=bins, labels=labels, right=False)

year_rating_crosstab = pd.crosstab(df['year'], df['rating_category'], margins=True, margins_name="Итого")
print("\nТаблица сопряженности для Год и Рейтинг:")
display(year_rating_crosstab)

# Рассчитываем долю каждой категории
year_rating_percentage = year_rating_crosstab.div(year_rating_crosstab.loc["Итого"], axis=1) * 100
print("\nДоля каждой категории (в %):")
display(year_rating_percentage)
```
